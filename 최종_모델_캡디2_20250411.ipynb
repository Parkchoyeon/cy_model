## pdf -> 텍스트 추출 & 전처리 & csv 저장

```python
import fitz  # PyMuPDF
import re
import pandas as pd

# 1. 전처리 함수 정의
def clean_text(text):
    text = re.sub(r"\d+\.\s*[가-힣a-zA-Z\d\s\?\!]+", "", text)  # 번호+제목 제거
    text = re.sub(r"([가-힣])([0-9])", r"\1 \2", text)  # 한글+숫자
    text = re.sub(r"([0-9])([가-힣])", r"\1 \2", text)
    text = re.sub(r"[^가-힣0-9\s]", " ", text)  # 특수문자 제거
    text = re.sub(r"\s+", " ", text).strip()  # 중복 공백 제거
    return text

# 2. PDF 텍스트 추출 함수
def extract_pdf_text(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""
    for page_num in range(len(doc)):
        if 50 <= page_num <= 52:  # <사례 1>~<사례 5>가 있는 페이지 제외
            continue
        full_text += doc[page_num].get_text()
    return full_text

# 3. 사례별 내담자/상담사 분리
def split_cases(text):
    # 각 사례 시작 구간 찾기 (번호. 제목)
    case_splits = re.split(r"\n\s*(\d+\.\s*[^\n<]+)\s*\n", text)
    cases = []

    # 첫 번째 요소는 머리말일 수 있어 무시
    for i in range(1, len(case_splits) - 1, 2):
        title = case_splits[i].strip()
        body = case_splits[i + 1].strip()

        # '<상담>' 기준으로 나누기
        if '<상담>' in body:
            prompt_part, response_part = body.split('<상담>', 1)
            prompt = clean_text(prompt_part)
            response = clean_text(response_part)
            cases.append({
                "prompt": f"내담자: {prompt}",
                "response": f"상담사: {response}"
            })

    return pd.DataFrame(cases)

# 4. 전체 파이프라인 실행
pdf_path = "/content/청소년 상담사례.pdf"
raw_text = extract_pdf_text(pdf_path)
df = split_cases(raw_text)

# 결과 저장
output_path = "/content/super_pdf_dataset1.csv"
df.to_csv(output_path, index=False, encoding='utf-8-sig')

import ace_tools as tools; tools.display_dataframe_to_user(name="전처리된 상담 사례 데이터", dataframe=df)
output_path
```

```python
def clean_text(text):
    # (1) '번호. 제목' 제거 (본문은 보존)
    text = re.sub(r"\d+\.\s*[가-힣a-zA-Z\d\s\?\!]+", "", text)

    # (2) 숫자+한글, 한글+숫자 분리
    text = re.sub(r"([가-힣])([0-9])", r"\1 \2", text)
    text = re.sub(r"([0-9])([가-힣])", r"\1 \2", text)

    # (3) 상담사 이름/소속/전화번호 제거 (ex. "석영미 청소년사업관 798 0325")
    text = re.sub(r"[가-힣]+\s+[가-힣]+\s*\d{2,4}\s*\d{3,4}", "", text)
    text = re.sub(r"[가-힣]+\s+[가-힣]+$", "", text)

    # (4) 특수문자 제거
    text = re.sub(r"[^가-힣0-9\s]", " ", text)

    # (5) 중복 공백 제거
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
import fitz  # PyMuPDF
import re
import pandas as pd
from tqdm import tqdm

# 1. 전처리 함수 정의
def clean_text(text):
    text = re.sub(r"\d+\.\s*[가-힣a-zA-Z\d\s\?\!]+", "", text)
    text = re.sub(r"([가-힣])([0-9])", r"\1 \2", text)
    text = re.sub(r"([0-9])([가-힣])", r"\1 \2", text)
    text = re.sub(r"[가-힣]+\s+[가-힣]+\s*\d{2,4}\s*\d{3,4}", "", text)
    text = re.sub(r"[가-힣]+\s+[가-힣]+$", "", text)
    text = re.sub(r"[^가-힣0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# 2. PDF에서 상담 사례 추출
def extract_dialogue_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""

    # 1) 목차 및 사례1~5 있는 페이지(0~12, 51~53) 제외
    exclude_pages = set(range(0, 13)) | set(range(50, 53 + 1))

    for page_num in range(len(doc)):
        if page_num in exclude_pages:
            continue
        full_text += doc[page_num].get_text()

    # 2) <상담> 기준으로 내담자 / 상담자 분리
    pattern = r"(\d+\.\s*[^\n<]+)\n(.*?)<상담>\n(.*?)(?=\n\d+\.|$)"
    matches = re.findall(pattern, full_text, re.DOTALL)

    data = []
    for _, client_text, counselor_text in tqdm(matches, desc="사례 추출 중"):
        client_clean = clean_text(client_text)
        counselor_clean = clean_text(counselor_text)

        if len(client_clean) < 1 or len(counselor_clean) < 1:
            continue

        data.append({
            "prompt": f"내담자: {client_clean}",
            "response": f"상담사: {counselor_clean}"
        })

    return pd.DataFrame(data)

# 3. 실행 및 저장
pdf_path = "/content/청소년 상담사례.pdf"
df = extract_dialogue_from_pdf(pdf_path)

# 저장
output_path = "/content/reset_pdf_dataset.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"✅ 목차 및 불필요 페이지 제거 포함, 총 {len(df)}건 저장됨 → {output_path}")
```

```python
# ✅ CounselGPT 모델에 청소년 상담 사례 추가 파인튜닝 전체 코드

import os
import pandas as pd
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    DataCollatorForLanguageModeling,
    TrainingArguments,
    Trainer,
)
from sklearn.model_selection import train_test_split

# 1. 디버깅 옵션 (필요 시)
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# 2. 전처리된 데이터 로딩
df = pd.read_csv("/content/gpt_pdf_response_cleaned.csv")
df = df[df['prompt'].notnull() & df['response'].notnull()]
df = df[df['response'].str.len() > 10]

# 3. 학습/검증 데이터 분할
train_df, val_df = train_test_split(df, test_size=0.1, random_state=42)
train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True))
val_dataset = Dataset.from_pandas(val_df.reset_index(drop=True))

# 4. 기존 파인튜닝된 CounselGPT 모델 불러오기
model_path = "/content/drive/MyDrive/CounselGPTModel"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)

tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id

# 5. 토크나이징 함수 정의
def tokenize(example):
    text = example["prompt"].strip() + "\n상담사: " + example["response"].strip()
    return tokenizer(text, truncation=True, padding="max_length", max_length=512)

# 6. 토크나이징 적용
train_dataset = train_dataset.map(tokenize)
val_dataset = val_dataset.map(tokenize)

# 7. 길이 기준 유효 샘플 필터링
def check_valid(example):
    return sum(example["attention_mask"]) > 10

train_dataset = train_dataset.filter(check_valid)
val_dataset = val_dataset.filter(check_valid)

print("✅ 학습 샘플 수:", len(train_dataset))
print("✅ 검증 샘플 수:", len(val_dataset))

# 8. 데이터 콜레이터
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False
)

# 9. 학습 인자 설정
training_args = TrainingArguments(
    output_dir="/content/students_counsel_refined",
    per_device_train_batch_size=4,
    num_train_epochs=2,
    save_steps=500,
    eval_steps=500,
    logging_steps=100,
    evaluation_strategy="steps",
    save_strategy="steps",
    save_total_limit=2,
    learning_rate=5e-5,
    fp16=torch.cuda.is_available(),
    seed=42,
)

# 10. 트레이너 정의
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
)

# 11. 학습 시작
trainer.train()

# 12. 모델 저장
model.save_pretrained("/content/drive/MyDrive/students_counsel_refined")
tokenizer.save_pretrained("/content/drive/MyDrive/students_counsel_refined")
print("✅ 추가 파인튜닝된 모델 저장 완료!")
```

## CounselGPT 모델 개선

```python
import os
import pandas as pd
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    DataCollatorForLanguageModeling,
    TrainingArguments,
    Trainer,
    pipeline
)
from sklearn.model_selection import train_test_split

# 디버깅 옵션
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# 1. 데이터 로딩
df = pd.read_csv("/content/final_prompt_response1.csv")
df = df[df["response"].notnull() & df["prompt"].notnull()]
df = df[df["response"].str.len() > 10]

# 2. 데이터 분할 (90% 학습 / 10% 검증)
train_df, val_df = train_test_split(df, test_size=0.1, random_state=42)
train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True))
val_dataset = Dataset.from_pandas(val_df.reset_index(drop=True))

# 3. 토크나이저 및 모델 불러오기
model_name = "skt/kogpt2-base-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name)
model.config.pad_token_id = tokenizer.pad_token_id
model.resize_token_embeddings(len(tokenizer))

# 4. 토크나이징 함수
def tokenize(example):
    text = example["prompt"].strip() + "\n상담사: " + example["response"].strip()
    return tokenizer(text, truncation=True, padding="max_length", max_length=768)

# 5. 토크나이징
train_dataset = train_dataset.map(tokenize)
val_dataset = val_dataset.map(tokenize)

# 6. 유효한 샘플만 필터링
def check_valid(example):
    return sum(example["attention_mask"]) > 10

train_dataset = train_dataset.filter(check_valid)
val_dataset = val_dataset.filter(check_valid)

print("✅ 학습 샘플 수:", len(train_dataset))
print("✅ 검증 샘플 수:", len(val_dataset))

# 7. 데이터 콜레이터
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False
)

# 8. 학습 설정
training_args = TrainingArguments(
    output_dir="/content/kogpt2_counsel_finetuned",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    save_steps=500,
    eval_steps=500,
    logging_steps=100,
    evaluation_strategy="steps",
    save_strategy="steps",
    save_total_limit=2,
    learning_rate=5e-5,
    fp16=torch.cuda.is_available(),
    seed=42,
)

# 9. 트레이너 정의
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
)

# 10. 학습
trainer.train()

# 11. 모델 저장
model.save_pretrained("/content/drive/MyDrive/kogpt2_counsel_finetuned")
tokenizer.save_pretrained("/content/drive/MyDrive/kogpt2_counsel_finetuned")
print("✅ 모델 저장 완료")

# 12. 테스트 (여러 문장)
chatbot = pipeline("text-generation", model=model, tokenizer=tokenizer, device=0 if torch.cuda.is_available() else -1)
test_prompts = [
    "내담자: 요즘 친구들과 다퉜는데 기분이 우울해요.\n상담사:",
    "내담자: 학업 스트레스로 너무 힘들어요.\n상담사:",
    "내담자: 가족과의 갈등이 심해서 지쳐요.\n상담사:",
    "내담자: 아무것도 하기 싫고 의욕이 없어요.\n상담사:",
    "내담자: 사람들과 어울리는 게 너무 힘들어요.\n상담사:"
]

for i, prompt in enumerate(test_prompts, 1):
    output = chatbot(prompt, max_new_tokens=150, do_sample=True, temperature=0.7, top_p=0.9, repetition_penalty=1.2)
    print(f"\n🧪 테스트 {i}:")
    print(output[0]['generated_text'])
```
